### Histogram


*   **`bins`**: Bins are the intervals into which the data is divided for counting and display in the histogram.
    - Having 20 bins means the data range will be split into 20 intervals
*   **`alpha`**: controls the transparency of the bars.
    - A value of 1.0 would be completely opaque, while 0.0 would be completely transparent.



```python
# Histogram - plot 1

rng= np.random.RandomState(12345)
plt.hist(rng.standard_normal(100), bins=20, color='blue', alpha= 0.7)
```

### **linestyle**
1. some other options for the linestyle argument:

*   `'-'` (**solid**): This is the default linestyle and creates a solid line.
*   `'--'` (**dashed**): Creates a dashed line.
*   `'-.'` (**dash-dot**): Creates a line with alternating dashes and dots.
*   `':'` (**dotted**): Creates a dotted line.
*   `'**None**'` or `' '` or `''` (no line): This will plot only markers if specified, or no line at all.

2. we can also use the shorter codes for these linestyles:
*   **`solid`**: `-`
*   **`dashed`**: `--`
*   **`dashdot`**: `-.`
*   **`dotted`**: `:`

```python
# Line Plot : plot 3
rng= np.random.RandomState(12345)
plt.plot(rng.standard_normal(50).cumsum(), color='blue', linestyle= 'dashed');

# cumsum cummulative summation
# Example:
# If the random numbers are [0.5, -0.2, 0.1, ...], 
# the cumulative sum becomes [0.5, 0.3, 0.4, ...].
```

### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots
3.   **Subplots** are numbered sequentially **starting from 1, going from left to right and top to bottom.**
    - a **(2, 2, 1)** argument means we're adding the first subplot (top-left corner) in the 2x2 grid.

```python
# create one or more subplots
# 2 x 2; 4 plots in total
# plot axis

rng = np.random.RandomState(12345)
fig = plt.figure(dpi=100)   # dpi dots  per inch BERAPA BESAR RAJAH
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.hist(rng.standard_normal(100), bins=20, color='orange', alpha=0.7);
ax2.scatter(np.arange(30), np.arange(30)+3*rng.standard_normal(30));
ax3.plot(rng.standard_normal(50).cumsum(), color='black', linestyle='dashed');
```

### **Adjusting the spacing around the subplots** 

* `hspace`= 0.5, height space
* `wspace`= 0.5, width space

```python
# An example that shrink the spacing of the subplots to zero
# Change the value of `hspace` and `wspace` 
rng = np.random.RandomState(12345)
fig, axes= plt.subplots(2,2, sharex=True, sharey=False)    # share x-axis, no sharing y-axis

# Using a nested loop to populate 2x2 grid
for i in range(2):
    for j in range(2):
        axes[i,j].hist(rng.standard_normal(500), bins=50, color='black', alpha=0.5)

# Adjusting the spacing
# fig.subplots_adjust(wspace= 0.5, hspace= 0.5)
```

### Setting the title, axis labels, ticks, and tick labels

```python
# An example
rng= np.random.RandomState(12345)
fig, ax= plt.subplots()
ax.plot(rng.standard_normal(1000).cumsum());

# Change x-axis ticks annotation
ax.set_xticks([0,250,500,750,1000]);

# Label the x-axis ticks with names
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five']);
ax.set_xlabel('Stages');
ax.set_ylabel('Value');
ax.set_title('My first matplotlib plot');
```

### ** Annotation and Drawing on a Subplot **

```python
# The SPX Data
from datetime import datetime

fig, ax= plt.subplots()

data= pd.read_csv('https://bit.ly/3iofIY8', index_col=0, parse_dates=True)
data
spx= data['SPX']
spx
spx.plot(ax=ax, color='blue')

crisis_data= [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')]

# label dalam graph (label, arrow)
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')
    
# zoom in on 2007-2010 xlimit, ylimit
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600,1800])
ax.set_title('Important dates in the 2008-2009 financial crisis');

# first column (index 0) of the CSV as the index of the DataFrame
# parse_dates=True, it will convert them into datetime objects
# ['SPX'] extract column SPX dari data
```

### **Adding shapes on the same plot**

```python
# Adding shapes to the same plot
fig, ax=plt.subplots()

# plt.rectangle((x,y), width, height, ..
rect= plt.Rectangle((0.2,0.75), 0.4, 0.15, color='black', alpha=0.3)

# circle ( centre, radius)
circ= plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)

# polygon every 3 edge
pgon= plt.Polygon([[0.15,0.15], [0.35, 0.4], [0.2,0.6]], color='green', alpha=0.5)

# shape akan appear bila guna .add_patch()
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon);
```

```python
# Challenge: plot a pentagon, ellipse, square figures in the exact same figure
fig, ax= plt.subplots()

# Pentagon
    # Center, sides, radius
pentagon= patches.RegularPolygon((0.2, 0.7), 5, radius=0.1, color='red')
ax.add_patch(pentagon);

# Ellipse
    # Center, width, height
ellipse= patches.Ellipse((0.7,0.7), 0.2, 0.1, color='blue')
ax.add_patch(ellipse);

# Square
    # lower-left, width, height
square= patches.Rectangle((0.4,0.2), 0.2, 0.2, color='green')
ax.add_patch(square);
```

```python
# create a frequency table
# 244 is the total whether it is horizontal or vertical
party_counts= pd.crosstab(tips['day'], tips['size'], margins=True)
party_counts
```

```python
# Reorder the index
party_counts= party_counts.reindex(index=['Thur', 'Fri', 'Sat', 'Sun'])
party_counts
```

```python
# Extract party_count from  column 2-5
# using loc()
# loc() is labelled based indexing
party_counts= party_counts.iloc[:, 1:5]

# indexing start from 0 exclusive column 5
```

```python
# show content for party_counts
party_counts
```

```python
# Normalize each row to sum to 1
party_pcts= party_counts.div(party_counts.sum(axis='columns'), axis='index')
party_pcts

# sum(axis=column) campur setiap column dlm 1 row=total
# axis= 'index' / 0 (column wise) divide setiap row /total 
```

```python
# Only select a few variables
# Using double bracket "[[]]"
data= macro[['cpi', 'm1', 'tbilrate', 'unemp']]
data
```

```python
print(data.isna().sum())
```

```python
# Applies a logarithm transformation
# Calculate the difference btw consecutive values
# Removes any rows with missing values (NaN)
trans_data= np.log(data).diff().dropna()
```

```python
# Generate scatter plot with regression line
fig= sns.regplot(x='m1', y='unemp', data=trans_data)

# Add a descriptive title that summarise the plot's content
fig.set(title='Changes in log(unemp) versus log (m1)');
```

```python
# kde = kernel desnity estimate - another name for density plot
# estimate of the continuous probability distribution
# one question for the final here
# pairplot-> a pair of y and x
# what is your observation? what is your recommendation?
sns.pairplot(trans_data, diag_kind='kde', kind='reg',
             plot_kws={'scatter_kws':{'alpha':0.2}});
```

a) CPI vs Other Variables
* CPI vs M1: Weak relationship, as the points are spread out.
* CPI vs Tbillrate: Appears negatively correlated (as CPI increases, Tbillrate tends to decrease).
* CPI vs Unemployment: Weak relationship; inflation and unemployment may not have a clear direct link.

b) M1 vs Other Variables
* M1 vs Tbillrate: Weak negative correlation, as seen in the downward trend.
* M1 vs Unemployment: No strong pattern, suggesting no clear link.

c) Tbillrate vs Other Variables
* Tbillrate vs Unemployment: Appears negatively correlated (higher Tbillrate, lower unemployment).

d) Unemployment vs Other Variables
* Unemployment vs CPI: Weak correlation.
* Unemployment vs Tbillrate: Negative correlation.

3. Key Takeaways
* Negative correlation between Tbillrate and CPI: Suggests that when interest rates rise, inflation decreases, which aligns with economic theory.
* Negative correlation between Tbillrate and Unemployment: Higher interest rates might be linked to lower unemployment, which can be unexpected (as higher rates usually slow economic growth).
* Weak relationship between M1 and other variables: Money supply may not directly affect other indicators in this dataset.
