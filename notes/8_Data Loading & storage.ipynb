## XML

```python
from lxml import etree
#create the root element
root=etree.Element("Book")

#add child element
book1=etree.SubElement(root,"book")
etree.SubElement(book1, "Title").text= "Phyton for Everybody"
etree.SubElement(book1, "Author").text= "John Doe"
etree.SubElement(book1, "Year").text= "2003"

book2=etree.SubElement(root, "book")
etree.SubElement(book2, "Title").text= "Phyton for Beginner"
etree.SubElement(book2, "Author").text= "Sar Williams"
etree.SubElement(book2, "Year").text= "2013"

xml_string= etree.tostring(root, pretty_print=True, xml_declaration=True, encoding="UTF-8")

#Write to file
with open('book.xml', 'wb') as file:
    file.write(xml_string)
```

```python
#read the XML file
tree=etree.parse('book.xml')

#extract the root
root=tree.getroot()
for elem in root.iter():
    print(elem.tag)
```

```python
#loop through each book element
for book in root.findall("book"):
    title=book.find("Title").text
    author=book.find("Author").text
    year=book.find("Year").text
    print(f"Title:{title}")
    print(f"Author:{author}")
    print(f"Year:{year}")
    print("---")
```

```python
#read performance file
tree=etree.parse('Performance_MNR.xml')

#extract the root
root=tree.getroot()
for elem in root.iter():
    print(elem.tag)
```

```python
#keluarkan data dari indicator 2nd row

for indicator in root.findall("INDICATOR"):
    name=indicator.find("AGENCY_NAME").text
    print(f"Name:{name}")
```

```python
from lxml import objectify
parsed=objectify.parse(open('Performance_MNR.xml'))
root=parsed.getroot()

data=[]
skip_fields=['PARENT_SEQ', 'INDICATOR_SEQ', 'DESIRED_CHANGE', 'DECIMAL_PLACEs']
for elt in root.INDICATOR:
    el_data={}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag]=child.pyval
    data.append(el_data)
```
